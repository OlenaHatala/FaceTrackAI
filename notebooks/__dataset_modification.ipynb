```python
import os
import shutil
import random
from glob import glob
from pathlib import Path
```

```python
directory = 'drive/MyDrive/FaceTrackAI/data/faces_nofaces/caltech-101'
folders = [f for f in os.listdir(directory) if os.path.isdir(os.path.join(directory, f))]
print(folders)
```

```python
base_dir = Path('drive/MyDrive/FaceTrackAI/data/faces_nofaces/caltech-101')
output_base = base_dir.parent

faces_dir = output_base / 'Faces'
nofaces_dir = output_base / 'NoFaces'

faces_dir.mkdir(exist_ok=True)
nofaces_dir.mkdir(exist_ok=True)

for folder in base_dir.iterdir():
    if not folder.is_dir():
        continue

    if folder.name == 'Faces':
        for img_path in folder.glob('*.jpg'):
            shutil.copy(img_path, faces_dir / img_path.name)
    else:
        image_paths = list(folder.glob('*.jpg'))
        sample_size = int(len(image_paths) * 0.1)
        sampled = random.sample(image_paths, sample_size)
        for img_path in sampled:
            target_path = nofaces_dir / f"{folder.name}_{img_path.name}"
            shutil.copy(img_path, target_path)

print("Done!")
```

```python
faces_path = 'drive/MyDrive/FaceTrackAI/data/faces_nofaces/Faces'
nofaces_path = 'drive/MyDrive/FaceTrackAI/data/faces_nofaces/NoFaces'

faces_count = len([f for f in os.listdir(faces_path) if os.path.isfile(os.path.join(faces_path, f))])
nofaces_count = len([f for f in os.listdir(nofaces_path) if os.path.isfile(os.path.join(nofaces_path, f))])

print(f"Faces: {faces_count} images")
print(f"NoFaces: {nofaces_count} images")
```
